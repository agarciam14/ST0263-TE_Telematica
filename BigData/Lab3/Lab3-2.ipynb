```pyspark
#import SparkSession
from pyspark.sql import SparkSession
```

```pyspark
#create spar session object
spark=SparkSession.builder.appName('data_processing').getOrCreate()
```

```pyspark
# 1. Cargar Datos 
df=spark.read.csv('s3://bigdata-telematica/telematica/datasets/spark/Casos_positivos_de_COVID-19_en_Colombia.csv',inferSchema=True,header=True)
df = df.withColumnRenamed(df.columns[0],"Fecha reporte web")\
.withColumnRenamed(df.columns[2],"Fecha de notificacion")\
.withColumnRenamed(df.columns[3],"Codigo divipola departamento")\
.withColumnRenamed(df.columns[5],"Codigo divipola municipio")\
.withColumnRenamed(df.columns[11],"Ubicacion del caso")\
.withColumnRenamed(df.columns[13],"Codigo iso del pais")\
.withColumnRenamed(df.columns[14],"Nombre del pais")\
.withColumnRenamed(df.columns[16],"Fecha de inicio de sintomas")\
.withColumnRenamed(df.columns[18],"Fecha de diagnostico")\
.withColumnRenamed(df.columns[19],"Fecha de recuperacion")\
.withColumnRenamed(df.columns[20],"Tipo de recuperacion")\
.withColumnRenamed(df.columns[21],"Pertenencia etnica")\
.withColumnRenamed(df.columns[22],"Nombre del grupo etnico")
```

```pyspark
# 2. Análisis exploratorio
#   2.1. Columnas
df.columns
```

```pyspark
#    2.2. Tipos de datos
df.printSchema()
```

```pyspark
#    2.3. Mostrar algunas columnas
df.select('Fecha reporte web','Fecha de notificacion','ID de caso','Edad','Estado').show(5)
```

```pyspark
#    2.4. Agregar columnas
df2 = df.withColumn("+18",((df["Edad"]>=18)&(df["Unidad de medida de edad"]==1)))
df2.printSchema()
```

```pyspark
#    2.5. Borrar columnas
df2 = df2.drop("+18")
df2.columns
```

```pyspark
#    2.6. Filtrar datos
df.filter((df["Edad"]>=18)&(df["Unidad de medida de edad"]==1)).select('Fecha reporte web','Fecha de notificacion','ID de caso','Edad','Unidad de medida de edad').show()
```

```pyspark
#    2.7 Ejecutar alguna función UDF o lambda sobre alguna columna creando una nueva.
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType,DoubleType,IntegerType

def unidades_edad(unidad):
    if unidad == 1:
        return 'anos'
    elif unidad == 2:
        return 'meses'
    else:
        return 'dias'
    
#create udf using python function
unidades_udf=udf(unidades_edad,StringType())
#apply udf on dataframe
df.withColumn('unidades_edad',unidades_udf(df['Unidad de medida de edad'])).select('Fecha reporte web','Fecha de notificacion','ID de caso','Edad','unidades_edad').show(10,False)
```

```pyspark
# 4. Guardar datos
#target directory 
write_uri='s3://bigdata-telematica/telematica/lab3-2/df_csv/'
```

```pyspark
# 3. contestar las siguientes preguntas sobre los datos de covid:
#    3.1 Los 10 departamentos con más casos de covid en Colombia ordenados de mayor a menor.
p31 = df.groupBy('Nombre departamento').count().orderBy('count',ascending=False).limit(10)
p31.coalesce(1).write.format("csv").option("header","true").save(write_uri+'4.1')
p31.show()
```

```pyspark
#    3.2 Las 10 ciudades con más casos de covid en Colombia ordenados de mayor a menor.
p32=df.groupBy('Nombre municipio').count().orderBy('count',ascending=False).limit(10)
p32.coalesce(1).write.format("csv").option("header","true").save(write_uri+'4.2')
p32.show()
```

```pyspark
#    3.3 Los 10 días con más casos de covid en Colombia ordenados de mayor a menor.
p33 = df.groupBy('Fecha reporte web').count().orderBy('count',ascending=False).limit(10)
p33.coalesce(1).write.format("csv").option("header","true").save(write_uri+'4.3')
p33.show()
```

```pyspark
#    3.4 Distribución de casos por edades de covid en Colombia.
p34 = df.groupBy('Unidad de medida de edad','Edad').count()
p34.coalesce(1).write.format("csv").option("header","true").save(write_uri+'4.4')
p34.show()
```

```pyspark
#    3.5 Realice la pregunda de negocio que quiera sobre los datos y respondala con la correspondiente programación en spark.
#        Distribución de casos por edades de covid en cada departamento de Colombia.
p35 = df.groupBy('Nombre departamento','Unidad de medida de edad','Edad').count().orderBy('Nombre departamento',ascending=True)
p35.coalesce(1).write.format("csv").option("header","true").save(write_uri+'4.5')
p35.show()
```
